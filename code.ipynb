```python
df['Review Rating'] = df.groupby('Category')['Review Rating'] \
                        .transform(lambda x: x.fillna(x.median()))
```

```python
df.columns = df.columns.str.lower().str.replace(' ', '_')
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels=labels)
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 90,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
pd.read_sql("PRAGMA table_info(customer);", conn)
```

```python
#Purchases by age group
pd.read_sql("""
SELECT age_group, COUNT(*) AS total_purchases
FROM customer
GROUP BY age_group
ORDER BY total_purchases DESC;
""", conn)
```

```python
#Most active locations:
pd.read_sql("""
SELECT location, COUNT(*) AS customer_count
FROM customer
GROUP BY location
ORDER BY customer_count DESC
LIMIT 10;
""", conn)
```

```python
#Highest repeat purchasers:
pd.read_sql("""
SELECT customer_id, previous_purchases
FROM customer
ORDER BY previous_purchases DESC
LIMIT 10;
""", conn)
```

```python
#Frequency vs spending:
pd.read_sql("""
SELECT
    purchase_frequency_days,
    AVG(purchase_amount) AS avg_spend
FROM customer
GROUP BY purchase_frequency_days
ORDER BY purchase_frequency_days;
""", conn)
```

```python
#Do discounts increase purchases?
pd.read_sql("""
SELECT discount_applied, COUNT(*) AS total_orders
FROM customer
GROUP BY discount_applied;
""", conn)
```

```python
#Revenue by season
pd.read_sql("""
SELECT season, SUM(purchase_amount) AS total_sales
FROM customer
GROUP BY season
ORDER BY total_sales DESC;
""", conn)
```

```python
#Top-performing categories:
pd.read_sql("""
SELECT category, SUM(purchase_amount) AS revenue
FROM customer
GROUP BY category
ORDER BY revenue DESC;
""", conn)
```

```python
#Revenue by Gender
pd.read_sql("""
SELECT gender, SUM(purchase_amount) AS revenue
FROM customer
GROUP BY gender;
""", conn)
```

```python
#Discount users who spent above avg
pd.read_sql("""
SELECT customer_id, purchase_amount
FROM customer
WHERE discount_applied = 'Yes'
AND purchase_amount >= (SELECT AVG(purchase_amount) FROM customer);
""", conn)
```

```python
#Top 5 products by avg review rating
pd.read_sql("""
SELECT item_purchased,
       ROUND(AVG(review_rating), 2) AS avg_rating
FROM customer
GROUP BY item_purchased
ORDER BY avg_rating DESC
LIMIT 5;
""", conn)
```

```python
#Compare shipping types
pd.read_sql("""
SELECT shipping_type,
       ROUND(AVG(purchase_amount), 2) AS avg_spend
FROM customer
GROUP BY shipping_type;
""", conn)
```

```python
#Do subscribed customers spend more?
pd.read_sql("""
SELECT subscription_status,
       COUNT(customer_id) AS total_customers,
       ROUND(AVG(purchase_amount),2) AS avg_spend,
       ROUND(SUM(purchase_amount),2) AS total_revenue
FROM customer
GROUP BY subscription_status
ORDER BY total_revenue DESC;
""", conn)
```

```python
#Highest discount usage per product
pd.read_sql("""
SELECT item_purchased,
       ROUND(
           100.0 * SUM(CASE WHEN discount_applied='Yes' THEN 1 ELSE 0 END)/COUNT(*),
       2) AS discount_rate
FROM customer
GROUP BY item_purchased
ORDER BY discount_rate DESC
LIMIT 5;
""", conn)
```

```python
#Segment customers (New, Returning, Loyal)
pd.read_sql("""
WITH customer_type AS (
    SELECT customer_id,
           previous_purchases,
           CASE
               WHEN previous_purchases = 1 THEN 'New'
               WHEN previous_purchases BETWEEN 2 AND 10 THEN 'Returning'
               ELSE 'Loyal'
           END AS customer_segment
    FROM customer
)
SELECT customer_segment, COUNT(*) AS num_customers
FROM customer_type
GROUP BY customer_segment;
""", conn)
```

```python
#Top 3 products per category
pd.read_sql("""
WITH item_counts AS (
    SELECT category,
           item_purchased,
           COUNT(customer_id) AS total_orders,
           ROW_NUMBER() OVER (
                PARTITION BY category
                ORDER BY COUNT(customer_id) DESC
           ) AS item_rank
    FROM customer
    GROUP BY category, item_purchased
)
SELECT item_rank, category, item_purchased, total_orders
FROM item_counts
WHERE item_rank <= 3;
""", conn)
```

```python
#Repeat buyers & subscription
pd.read_sql("""
SELECT subscription_status,
       COUNT(customer_id) AS repeat_buyers
FROM customer
WHERE previous_purchases > 5
GROUP BY subscription_status;
""", conn)
```

```python
#Revenue by age group
pd.read_sql("""
SELECT age_group,
       SUM(purchase_amount) AS total_revenue
FROM customer
GROUP BY age_group
ORDER BY total_revenue DESC;
""", conn)
```
